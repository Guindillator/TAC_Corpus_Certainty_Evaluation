# 95% Confidence Interval for Proportions in Survey 1

```r
# Set CI alpha level (1-alpha/2)*100%
alpha = 0.05

# Load Data with proportion
HIGH = Agreement_s1$"PO-High"
MED_HIGH = Agreement_s1$"PO-Medium High"
MED_LOW = Agreement_s1$"PO-Medium Low"
LOW = Agreement_s1$"PO-Low"

# Calculate the critical z-score
z = qnorm(1-alpha/2)

# Compute the CI
for (x in 1:45){
#h_lower <- HIGH[x]+c(+1,-1) *z*sqrt(HIGH[x]*(1-HIGH[x])/sum(Dataset_Merge_s1[x,2:5]))
h_upper <- HIGH[x] +c(+1,-1)*z*sqrt(HIGH[x]*(1-HIGH[x])/sum(Dataset[x,1:4]))
#mh_lower <- MED_HIGH[x] +c(+1,-1)*z*sqrt(MED_HIGH[x]*(1-MED_HIGH[x])/sum(Dataset_Merge_s1[x,2:5]))
mh_upper <- MED_HIGH[x]+ c(+1,-1)*z*sqrt(MED_HIGH[x]*(1-MED_HIGH[x])/sum(Dataset[x,1:4]))
#ml_lower <- MED_LOW[x] c(+1,-1)*z*sqrt(MED_LOW[x]*(1-MED_LOW[x])/sum(Dataset_Merge_s1[x,2:5]))
ml_upper <- MED_LOW[x]+ c(+1,-1)*z*sqrt(MED_LOW[x]*(1-MED_LOW[x])/sum(Dataset[x,1:4]))
#l_lower <- LOW[x] +c(+1,-1)*z*sqrt(LOW[x]*(1-LOW[x])/sum(Dataset_Merge_s1[x,2:5]))
l_upper <- LOW[x] +c(+1,-1)*z*sqrt(LOW[x]*(1-LOW[x])/sum(Dataset[x,1:4]))
    
tiff(file = paste(x, '.tiff', sep = ''), width = 3200, height = 3200, units = "px", res = 800)
boxplot(h_upper,mh_upper, ml_upper,l_upper,names = c("High", "Medium High", "Medium Low", "Low"), xlab="Categories", ylab="95% Confidence Interval", main= paste("Statement ", x, sep=''),
        col=topo.colors(4))
dev.off()
boxplot(h_upper,mh_upper, ml_upper,l_upper, names = c("High", "Medium High", "Medium Low", "Low"), xlab="Categories", ylab="95% Confidence Interval", main= paste("Statement ", x, sep=''), 
        col=topo.colors(4))}
```
